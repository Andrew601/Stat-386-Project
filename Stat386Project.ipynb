```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup


driver = webdriver.Chrome() 
url = 'https://www.basketball-reference.com/players/shooting.fcgi?player_id=embiijo01&year_id=2023&is_playoffs=0'
driver.get(url)

soup = BeautifulSoup(driver.page_source, 'html.parser')

data_list = []

tooltip_divs = soup.find_all('div', class_='tooltip')

# Iterate through each tooltip element and extract the required information
for tooltip_div in tooltip_divs:
    tip_data = tooltip_div['tip'].split('<br>')

    # Extracting relevant information
    date = tip_data[0].split(", ")[0]
    quarter = tip_data[1]
    action = tip_data[2]
    top = tooltip_div['style'].split(';')[0].split(':')[1].replace('px', '').strip()
    left = tooltip_div['style'].split(';')[1].split(':')[1].replace('px', '').strip()
    result = tooltip_div['class'][1]

    # Create a dictionary
    entry = {
        'Date': date,
        'Quarter': quarter,
        'Action': action,
        'Top': top,
        'Left': left,
        'Result': result
    }

    data_list.append(entry)

driver.quit()

# Convert the list of dictionaries to a DataFrame
df = pd.DataFrame(data_list)

df['Player'] = 'Joel Embiid'
df['Quarter'] = df['Quarter'].str.extract('(\d)')
df['Distance (ft)'] = df['Action'].str.extract('(\d+) ft').astype(float)
df['Top'] = df['Top'].astype(int)
df['Left'] = df['Left'].astype(int)

df.to_csv('JoelEmbiid.csv', index=False)

```
